```python
status_distribution = pd.DataFrame({
    'count': orders['order_status'].value_counts(),
    'percentual': orders['order_status'].value_counts(normalize=True)*100
}).round(2)
status_count = orders['order_status'].value_counts()
ax = status_count.plot(kind='bar')
for i, v in enumerate (status_count):
    ax.text(i,v +50,str(v),ha='center')
status_distribution
```
